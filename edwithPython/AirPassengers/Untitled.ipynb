```python
import pandas as pd
import numpy as np
```

```python
df_time_series = pd.read_csv("./AirPassengers.csv")
df_time_series.head()
```

```python
df_time_series["step"] = range(len(df_time_series))
df_time_series["cum_sum"] = df_time_series["#Passengers"].cumsum() # 로우를 계속 더해감
df_time_series["cum_max"] = df_time_series["#Passengers"].cummax() # 로우당 최고점
df_time_series["cum_min"] = df_time_series["#Passengers"].cummin() # 최저점
df_time_series.head()
```

```python
temp_date = df_time_series["Month"].map(lambda x: x.split("-"))
temp_date = np.array(temp_date.values.tolist())
temp_date[:5]
```

```python
df_time_series["year"] = temp_date[:, 0]
df_time_series["month"] = temp_date[:, 1]
df_time_series.head()
```

```python
df_time_series["diff"] = df_time_series["#Passengers"].diff().fillna(0)
```
